# シンプソンの公式の導出
---

$f(x)$上に3点をとる。両端を$a$、$b$とし、中点を$c$とする。  
$f(x)$をこの3点を通る放物線で近似をし、$\int_{a}^{b}f(x)dx$の近似値を求める。

$(a, f(a))$を通る2次関数 (点$b$、点$c$でもよい)

$$
y = \alpha (x-a)^{2} + \beta (x-a) + f(a)
$$

これが$(b, f(b)), (c, f(c))$を通るので、

$$
\begin{align}
f(b) &=& \alpha (b-a)^{2} + \beta (b-a) + f(a) \\
f(c) &=& \alpha (c-a)^{2} + \beta (c-a) + f(a) \\
c-a &=& \frac{a+b}{2} - a = \frac{a+b-2a}{2} = \frac{b-a}{2}
\end{align}
$$

(3)を(2)に代入

$$
f(c) = \frac{\alpha}{4} (b-a)^2 + \frac{\beta}{2} (b-a) + f(a) \quad (2')
$$

$\alpha$と$\beta$を求める

$(1)-(2')\times2$より

$$
\alpha = \frac{2}{(b-a)^{2}}\{f(a) + f(b) - 2f(c)\}
$$

$(1)-(2')\times4$より

$$
\beta = \frac{1}{b-a}\{-3f(a)-f(b)+4f(c)\}
$$

$f(x)$を2次関数で近似をして積分をするので、

$$
\begin{align}
\int_{a}^{b}f(x)dx &\approx& \int_{a}^{b}\{\alpha(x-a)^{2}+\beta(x-a)+f(a)\}dx \nonumber \\
&=& [\frac{\alpha}{3}(x-a)^{3}+\frac{\beta}{2}(x-a)^{2}+f(a)x]_{a}^{b} \nonumber \\
&=& \frac{\alpha}{3}(b-a)^{3}+\frac{\beta}{2}(b-a)^{2}+f(a)(b-a) \nonumber \\
&=& \frac{1}{3}\frac{2}{(b-a)^{2}}\{f(a)+f(b)-2f(c)\}(b-a)^3+\frac{1}{2}\frac{1}{b-a}\{-3f(a)-f(b)+4f(c)\}(b-a)^{2}+f(a)(b-a) \nonumber \\
&=& (b-a)\{\frac{2}{3}(f(a)+f(b)-2f(c))+\frac{1}{2}(-3f(a)-f(b)+4f(c))+f(a)\} \nonumber \\
&=& \frac{b-a}{6}(4f(a)+4f(b)-8f(c)-9f(a)-3f(b)+12f(c)+6f(a)) \nonumber \\
&=& \frac{b-a}{6} (f(a)+4f(c)+f(b)) \nonumber
\end{align}
$$

よって、
$$
\int_{a}^{b}f(x)dx \approx \frac{b-a}{6}\{f(a)+4f(\frac{a+b}{2})+f(b)\}
$$
